## Demo: How to work with database


```python
import requests

# URL of the file to download
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Save the content to a file
    with open("m4_survey_data.sqlite", "wb") as f:
        f.write(response.content)
    print("File downloaded successfully!")
else:
    print("Failed to download file:", response.status_code)
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
df_all = pd.read_sql_query("SELECT * FROM master", conn)
plt.hist(df_all['ConvertedComp'], bins=10)  # Adjust the number of bins as needed
plt.xlabel('ConvertedComp')
plt.ylabel('Frequency')
plt.title('Histogram of ConvertedComp')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
plt.boxplot(df_all['Age'].dropna())
plt.ylabel('Age')
plt.title('Box Plot of Age with Nulls Excluded')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

plt.scatter(df_all['Age'], df_all['WorkWeekHrs'])
plt.xlabel('Age')
plt.ylabel('Hours Worked / Week')
plt.title('Raw Hours')
plt.show()
```

```python
possible_Hours_only = df_all[df_all['WorkWeekHrs'] <= 168]
```

```python
plt.scatter(possible_Hours_only['Age'], possible_Hours_only['WorkWeekHrs'])
plt.xlabel('Age')
plt.ylabel('Hours Worked / Week')
plt.title('Non-sensical Hours Excluded')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
df_all['CodeRevHrs'].describe()
```

```python
# your code goes here
plt.figure(figsize=(12, 8))
plt.scatter(possible_Hours_only['WorkWeekHrs'], possible_Hours_only['CodeRevHrs'], s= possible_Hours_only['Age'])  # 's' parameter specifies the size of the bubbles
plt.xlabel('Hours Worked / Week (Exluding Non-Sensical)')
plt.ylabel('Hours Reviewing Code / Week')
plt.title('Bubble Plot of Hours Worked vs Hours Reviewing Code (Bubble Size Relates to Age)')
plt.show()
conn.close()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
conn = sqlite3.connect("m4_survey_data.sqlite")
```

```python
df_DatabaseWorkedWith = pd.read_sql_query("""
SELECT *
FROM DatabaseWorkedWith
""", conn)

respondent_counts = df_DatabaseWorkedWith.groupby('Respondent').size()

one_response = respondent_counts[respondent_counts == 1].index.tolist()

one_database_used = df_DatabaseWorkedWith[df_DatabaseWorkedWith['Respondent'].isin(one_response)]
mysql_only = one_database_used[one_database_used['DatabaseWorkedWith'] == 'MySQL']
print("Respondents only using MYSQL:" ,len(mysql_only))
```

```python
df_LanguageDesireNextYear = pd.read_sql_query("""
SELECT LanguageDesireNextYear, Count(*) as Count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY Count DESC
LIMIT 5
""", conn)
plt.figure(figsize=(8, 6))
plt.pie(df_LanguageDesireNextYear["Count"], labels=df_LanguageDesireNextYear["LanguageDesireNextYear"], autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Top 5 Languages to Learn Next Year')  
plt.show()
```

```python
# your code goes here
df_DatabaseDesireNextYear = pd.read_sql_query("""
SELECT DatabaseDesireNextYear, Count(*) as Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
""", conn)
plt.figure(figsize=(8, 6))
plt.pie(df_DatabaseDesireNextYear["Count"], labels=df_DatabaseDesireNextYear["DatabaseDesireNextYear"], autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Top 5 Databases to Learn Next Year')  
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
hours_distribution_by_age = df_all.groupby(['Age'])[list(['WorkWeekHrs', 'CodeRevHrs'])].median()
hours_distribution_by_age_df = pd.DataFrame(hours_distribution_by_age)
df_stacked = hours_distribution_by_age_df.iloc[14:20]

fig, ax = plt.subplots()

ax.bar(df_stacked.index, df_stacked['WorkWeekHrs'], label='Work Week Hours')

ax.bar(df_stacked.index, df_stacked['CodeRevHrs'], bottom=df_stacked['WorkWeekHrs'], label='Code Review Hours')

ax.set_xlabel('Age')
ax.set_ylabel('Hours')
ax.set_title('Weekly Hours Worked and Spent in Code Review for Ages: 30 - 35')
ax.legend()

plt.show()
```
